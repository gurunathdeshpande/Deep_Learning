**Removing the stopwords**

```python
import nltk 
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = stopwords.words('english')
print(stop_words)
sent = "India is republic nation We are proud Indians . We are Love India . India India India . Jai Hind ."
word_tokens = word_tokenize(sent)
filtered_sentence = [w for w in word_tokens if w.lower() not in stop_words]
print(word_tokens)
print(filtered_sentence)
```

**Stemmer**

```python
from nltk.stem.snowball import SnowballStemmer
stemobj = SnowballStemmer('english')
stemobj.stem('Walked')
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem.snowball import SnowballStemmer
sent = "makeing walked running runned"
word_tokens=word_tokenize(sent)
stemmed_vocab = []
stemobj = SnowballStemmer("english")
for i in word_tokens:
  stemmed_vocab.append(stemobj.stem(i))
print(stemmed_vocab)
```

**Lemmatization**

```python
from nltk.stem.wordnet import WordNetLemmatizer
lemmaobj=WordNetLemmatizer()
lemmaobj.lemmatize("went",pos='v')
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem.wordnet import WordNetLemmatizer
sent="went gone running ran playing played are is"
word_tokens=word_tokenize(sent)
lemma_vocab=[]
lemmaobj=WordNetLemmatizer()
for i in word_tokens:
    lemma_vocab.append(lemmaobj.lemmatize(i,pos='v'))
print(lemma_vocab)
```

**Dropping stop words**

```python
import spacy
#loading the english language small model of spacy
en = spacy.load('en_core_web_sm')
stopwords = en.Defaults.stop_words
text = " we will show how to remove stopwords using spacy library"
lst=[]
for token in text.split():
 if token.lower() not in stopwords:
   lst.append(token)

#Join items in the list
print("Original text : ",text)
print("Text after removing stopwords : ",' '.join(lst))
```
